**Functions**

```python
def hi(name):
  return f"Hello {name}!!"
print(hi("Students"))
```

**Function with default arguments**

```python
def sample(x,y=1):
  return x*y
print(sample(5))
print(sample(5,3))
```

Function with keyword arguments  "*arg"

```python
def result(*arg):
  return sum(arg)
print(result(1,2,3))
print(result(11,22,33,44))
```

**Function with Kwargs**

```python
def info(**kwargs):
  for key,value in kwargs.items():
    print(f"{key}:{value}")
info(name="Abhi",age=20,city="vijayawada")
```

**Lambda functions**

```python
sq=lambda x:x*x
print(sq(5))
add=lambda x,y:x+y
print(add(3,5))
power=lambda x,y:x**y
print(power(2,5))
pq=lambda a,b:a^b
print(pq(5,6))
```

**Write a program to return n number of multiple values where passing a list of 1 to 5 elements**

```python
def nums(a):
  return min(a),max(a),sum(a)/len(a)
min_val,max_val,avg=nums([1,2,3,4,5])
print(f"min:{min_val},max:{max_val},avg:{avg}")
```

**RECURSION**

**DIRECT: Recurisive factorial**

```python
def fact(n):
  if n==0 or n==1:
    return 1
  return n*fact(n-1)
print(fact(5))
```

**INDIRECT: Even or Odd**

```python
def e(n):
  if n==0:
    return True
  return o(n-1)
def o(n):
  if n==0:
    return False
  return e(n-1)
print(e(2))
print(o(3))
```

**TAIL :FACT**

```python
def fact(n,t=1):
  if n==0 or n==1:
    return t
  return fact(n-1,n*t)
print(fact(5))
```

**LINEAR**

```python
def total(n):
  if n<0:
    return 0
  return n+total(n-1)
print(total(10))
```

**NESTED** **SUPER FACTORIAL**

```python
def sfact(n):
  if n<=0:
    return 1
  return fact(n)*sfact(n-1)
def fact(n):
  if n<=1:
    return 1
  return n*fact(n-1)
print(sfact(4))
print(sfact(3))
```

**POWER TOWER**

```python
def pt(a,n):
  if n==1:
    return a
  return a**pt(a,n-1)
print(pt(2,4))
print(pt(3,2))
```
